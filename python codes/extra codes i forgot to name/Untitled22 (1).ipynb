```python
import numpy as np
from sympy import sinh
import matplotlib.pyplot as plt
import seaborn as sn


x = np.linspace(0, 1, 100)

def y_1(G,x):
  zeta_1 = -1
  zeta_2 = -1
  psi = (-1) * np.sinh(20)**(-1) * (zeta_1*np.sinh(20*(1-x)) + zeta_2*np.sinh(20*x))
  y_1 = G*x*(1-x)-zeta_1*(1-(psi/zeta_1)) + x*(zeta_1 - zeta_2)
  return y_1

m = y_1(5,x)
a = y_1(3,0)
g = y_1(0,x)
i = y_1(-3,x)
b = y_1(-5,x)


plt.plot(x, m, label='kr = 5')
plt.plot(x, a, label='kr = 3')
plt.plot(x, g, label='kr = 0')
plt.plot(x, i, label='kr = -3')
plt.plot(x, b, label='kr = -5')





#j = y_1(0,x)

G_value =[5,3,0,-3,-5]

for G in zip(G_value):
    y_values = y_1(G,x)
    plt.plot(x, y_values, label=f'$G={G}$')

#plt.xlabel('$q^2 (GeV^2)$')
#plt.ylabel('$d\Gamma/dq^2/|V_{cb}|^2 ps^{-1}GeV^{-2}$')
#plt.xlim(8, 9.2)
plt.legend()
#plt.figure(figsize=(10, 6))  # Adjust figure size as needed
#plt.show()
#plt.plot(x, j, label='kr = 20')
#plt.xlim(0,0.6)
#plt.ylim(0,2)
plt.xlabel('y')
plt.ylabel('$\Psi(y)$')
#plt.title('Plot of y(x), y_1(x), and y_2(x)')
plt.legend()
plt.grid(True)
plt.show()
```
